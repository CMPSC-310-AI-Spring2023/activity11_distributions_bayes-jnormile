## Binomial Distribution

The plot provided by the code below demonstrates the results of running 100 trials that each involve 100 individual tests. Given that the odds of success/failure are equivalent at `0.5` each, these tests can be thought of as little more than simple coin flips. This means that the numeric value associated with each trial is the number of "successes" wherein a success could be compared to flipping heads.

Given these parameters, we expect *about* 50 successful flips for each trial. True to form, most of the generated trials are within 10 coin flips of 50, with just one notable outlier only hitting 33 successes (this outlier can be *immediately* identified in the plot below).

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

This segment of code generates `100` random values between `1` and `10`. Wherein the prior distribution saw our plot points bunched around a particular value (50), in this distribution we saw no rhyme or reason in the scattered plot points--indeed, the provided plot appears to chaotically fluctuate from one value to the next.

Since the values are randomly generated, this intuitively follows reason.

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

The normal distribution below sees a "bunched" up distribution of plot points between `0` and `0.2`. If we were to chart the number of occurrences of these values, I imagine the resulting plot would look like an umbrella curve.

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
